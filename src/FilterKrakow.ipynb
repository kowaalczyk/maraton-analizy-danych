```python
import pandas as pd


def strip_string(s):
    if type(s)==str:
        s = s.strip()
    return s


def clean(df):
    df = df.iloc[1:]
    df = df.dropna(how='any')
    df = df.rename(columns=lambda x: x.strip())
    df = df.applymap(strip_string)
    return df
```
